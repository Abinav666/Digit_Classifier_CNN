```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2 as cv
```

```python
# Importing Datasets
train=pd.read_csv('/content/drive/MyDrive/MNIST_Project/mnist_train.csv').values
test=pd.read_csv('/content/drive/MyDrive/MNIST_Project/mnist_test.csv').values
```

```python
# Perprocessing training data
y_train=train[:,:1]
X_train=train[:,1:]
X_train=X_train.reshape(60000,28,28,1)
X_train=X_train/255.0
```

```python
# Perprocessing testing data
y_test=test[:,:1]
X_test=test[:,1:]
X_test=X_test.reshape(10000,28,28,1)
X_test=X_test/255.0
```

CNN MODEL BUILDING

```python
# Importing ANN & CNN libraries
# Dense is used to create one dimensional layer of neurons
# Conv2D invokes to convolutional principles
# Dropout is going to cut down the number of connections between the neuron
# Flatten is used to convert the 2D to a 1D list
# Max pooling will take the maximum out of the divided convolved image 
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Dropout, MaxPooling2D, Flatten
```

```python
# Making "model" to be the sequential()
model=Sequential()
```

```python
# Making my Convolutional Neural Network

model.add(Conv2D(28,kernel_size=(3,3), input_shape = (28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128,activation=tf.nn.relu))
model.add(Dropout(0.2))

model.add(Dense(10,tf.nn.softmax))
```

```python
# Compiling the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#Training the model
model.fit(X_train,y_train,epochs=10)
```

```python
model.evaluate(X_test,y_test)
```
